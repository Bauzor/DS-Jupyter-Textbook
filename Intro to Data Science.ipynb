<h1>1.1 What is Data Science?</h1>
<p>Data Science is an emerging field of technology in current society where the demand grows as the amount of information that people are collecting is growing. This information is not always easy to read and understand. Take this Titanic data set as an example:</p>

<img src = "images/train_csv.jpg">

<p>The data set above is ugly and hard to work with. The job of a data scientist is to turn this data into something more user-friendly so that we can draw some conclusions and answer certain questions. This is done by taking information and performing statistical analysis upon the information and then applying computer science algorithms to process, visualize and predict with said data. However, note that a data scientist may or may not be familiar with the field of information the data scientist is working with and thus a specialist of the subject is generally important when it comes to making certain decisions. Thus, as a data scientist, it is important that we are always open to learning about a variety of subjects from determining what song to suggest a user, to predicting housing prices.</p>

<h1>What is our goal?</h1>
<p>We will begin by determining our goal with our data set and then splitting this up into smaller and smaller questions that will highlight exactly what tasks we are going to perform on our dataset to achieve set goal. Given our data set, the goal that I have come up with is to create an analysis and find out what factors affect whether or not a passenger with an unknown outcome will survive or die on the titanic. When setting our target -<i>the target of our data is the feature/variable that we are trying to predict</i>- it is important to note whether or not we have or are able to collect information on our target.</p>
<h2>To outline again, our goal is to predict whether or not a passenger will survive or die aboard the titanic. We can then break this question down into a bunch of smaller problems</h2>
<ul>
    <li>How can we mung and wrangle data so that our data is clean eough to work with?</li>
    <li>What hypotheses do we want to make about certain variables and it's correlation with other variables?</li>
    <li>How can we visualize our data in such a way that we can communicate and justify certain decisions we will make with our data?</li>
    <li>How can we use our chosen variables to model a system to make predictions?</li>
    <li>How close were our predictions on certain tests?</li>
    <li>How can we improve our workflow for our next data set analysis?</li>
</ul>
<p>Breaking each part down even further:</p>
<ol>
    <li>When setting a goal, always ask sharp questions, not vague</li>
    <ul>
        <li>Sharp questions are questions that are extremely specfic such that there is a clear answer. Here are a few examples of good questions to ask:
            <ul>
                <li>Will my passenger survive aboard the titanic based on the data set of people survived and died?</li>
                <li>Will I enjoy this song based on all the other music that I've listened to and other people who listen to similar music as myself</li>
            </ul>
        </li>
        <li>Examples of vague questions that are not good questions to ask as a data scientist include:
            <ul>
                <li>Will the price of this stock go up?</li>
                <li>What can my data tell me about the titanic?</li>
            </ul>
        </li>
        <li>All vague questions can be split into sharp questions:
            <ul>
                <li>Will the price of this stock go up?
                    <ul>
                        <li>How many days will it take for my stock price to go up by?</li>
                        <li>How much will my stock price go back up for and how long will it stay up?</li>
                    </ul>
                </li>
                <li>What can my data tell me about the titanic?
                    <ul>
                        <li>What factors/features affect the survival </li>
                        <li>In the movie women and children boarded life boats first, Was this really in the case in real life?
                        </li>
                    </ul>  
                </li>
            </ul>
        </li>
    </ul>
    <li>How can we mung our data so that it is clean enough to work with?</li>
    <ul>
        <li>
            Putting our data into a table
            <ul>
                <li>
                    Most python data scientists use the Pandas library to organize data into objects called dataframes. 
                </li>
                <ul>
                    <li>A dataframe is an object that is very similar to what an excel document looks like: rows and columns. A 
                        unit of a data frame is a series which is a column of a dataframe and is similar to a dictionary.
                    </li>
                </ul>
            </ul>
        </li>
        <li>Clean our data
                    <ul>
            <li>In reality, we often get data that has a lot of missing values, it is important to check for this before trying 
                manipulate and use the data. There are many methods for dealing with missing values. Generally it is better to 
                try and fill in the missing data than to omit the entire entry due to one line of missing information.
            </li>
            <ul>
                <li>Find the source of the data and see if they have the answers you're seeking for
                </li>
                <li>Research online to see if you can fill in the data yourself</li>
                <li>Making a guess using known information. For example, if we were working with numerical data we could, take 
                    the average value of that column and fill in the rest of the columns with that value
                </li>
            </ul>
            <li>Data Quality is extremely important.
                <ul>
                    <li>If we are given inconsistent or poor column header names, we can rename the column headers.
                    </li>
                    <li>data format can affect how we do analysis. If in birth date column we are given the string "sept 27 
                        2017" 
                        and 07/01/2017 in the same column we would want to convert it into a consistent format.
                    </li>
                    <li>After, it would probably also be easier to work with the age of the person rather than their date of 
                        birth
                        so we can also calculate that into years and months
                    </li>
                    <li>it may also be better to only work with one unit so converting the months into decimal values in the 
                        year
                        unit would be better to work with
                    </li>
                    <li> there are many histogram graph functions that are built in that will help us see what types of values 
                        we

                    </li>
                </ul>
            </li>
        </ul>
        </li>
    </ul>
    <li>Feature Engineering and data visualizaiton for decision justification and communication</li>
    <ul>
        <li>Feature Engineering is the act of taking variables provided from the dataset and transforming them into direct
            variables that are used for training our models, or coming up with a conclusion
        </li>
        <li>a common starting point is simply plotting every variable against eachother into a graph that is known as a 
            correlation plot.
        </li>
        <li>The coefficient of determination is another factor that helps determine how well a variable predicts another.
            It ranges from 0 to 1 with 1 being a strong direct correlation and 0 being no correlation.
        </li>
        <li>Example of feature engineering: consider the three variables, speed of train, departure time of train, and arrival 
            time of train. If we plot both arrival against speed, or departure against speed there will be little to no 
            correlation. If we were to subtract arrival time from departure time we would get the amount of time it took for the 
            train to travel from point a to point b. Then we could simply form a regression model and to interpolate or 
            extrapolate how fast a train were travelling depending on a new departure time and arrival time. Interpolate means 
            to make a prediction within the graph of what our data covers, and extrapolate means to make a prediction outside
            of what our data covers.
        </li>
        <li>There are many commonly used feature engineering tricks for different types of data</li>
        <ul>
            <li>Scale-Invariant Feature Transform(SIFT) is an image recognition type of feature engineering
            </li>
            <li>Text-frequency Inverse document frequency (TF-IDF) determines how important a word is within its document. 
                Various text recommender systems use this to suggest words that you use more commonly.
            </li>
            <li>For specific fields there are known formula calculations. For example if we were in economics we could collect 
                data on the number of employed, unemployed, the labour force and compute a ratio known as the labour - force 
                participation rate and then use that to predict for something
            </li>
            <li>Deep learning can be use to find features by means of clustering
            </li>
        </ul>
    </ul>
    <li>Modelling</li>
    <ul>
        <li>After we have all the needed features we must choose appropriate models to train and use for our problem at 
            hand. Common problems that people ask can be answered with specfic models
        </li>
        <ul>
            <li>How many/much: regression model (numerical supervised learning)</li>
            <li>What category: classification model (categorical supervised learning)</li>
            <li>Which group: cluster model (unsupervised learning)</li>
            <li>is it weird: custom model that analyzes purchase patterns (can be a combination of supervised unsupervised and reinforcement</li> 
            <li>which action: reinforcement learning model</li>
        </ul>
    </ul>
    <li>Workflow improvement and Error</li>
    <ul>
        <li>partition the data in multiple ways using random library and train it with a different amount of data and see
            how well it performs on the data that you already have on average. Usually 60/40 partitions or 70/30 partitions are
            good.
        </li>
        <li>tweak different values, models or features to get better predictions and then when satisfied, train a 100% of the
            data and use it to make predictions
        </li>
        <li>take notes on how well you score and whenver there is new data, always add it and update the models to get better
            predictions
        </li>
        <li>remember that when you make conclusions, there is a difference between causation and correlation</li>
    </ul>
</ol>
